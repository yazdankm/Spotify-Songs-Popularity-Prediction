```python
# Finding the top ten songs with popularity more than 75
from tabulate import tabulate
import pandas as pd

# Assuming 'df' is your DataFrame
df['track_popularity'] = pd.to_numeric(df['track_popularity'], errors='coerce')

# Drop rows with NaN popularity
df = df.dropna(subset=['track_popularity'])

# Filter songs with popularity greater than 75 and select top ten songs
filtered_songs = df[df['track_popularity'] > 75]
print("Filtered Songs:", len(filtered_songs))

# Select top ten songs
top_ten_songs = filtered_songs.nlargest(10, 'track_popularity')
print("Top Ten Songs:", len(top_ten_songs))

# Display the result in a more presentable format
table = tabulate(top_ten_songs[['track_name', 'track_artist', 'track_popularity']],
                 headers={'track_name': 'Track Name', 'track_artist': 'Artist', 'track_popularity': 'Popularity'},
                 showindex=False, tablefmt='pretty')

# Print the table
print(table)
```

```python
#Seasonality: Are there seasonal or yearly trends in track popularly, tempo, valence.(question#3)

#Convert Release Date to DateTime:
df['track_album_release_date'] = pd.to_datetime(df['track_album_release_date'])
```

```python
#Extract the year and month from the release date to facilitate analysis:
df['release_year'] = df['track_album_release_date'].dt.year
df['release_month'] = df['track_album_release_date'].dt.month
df.head(5)
```

```python
#Group Data by Year and Month:
seasonal_data = df.groupby(['release_year', 'release_month']).agg({
    'track_popularity': 'mean',
    'tempo': 'mean',
    'valence': 'mean'
}).reset_index()
```

```python
#Plot line charts or heatmaps to visualize the seasonal trends in popularity, tempo, and valence:
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 10))

# Popularity over time
plt.subplot(3, 1, 1)
sns.lineplot(x='release_month', y='track_popularity', hue='release_year', data=seasonal_data)
plt.title('Popularity Over Time')

# Tempo over time
plt.subplot(3, 1, 2)
sns.lineplot(x='release_month', y='tempo', hue='release_year', data=seasonal_data)
plt.title('Tempo Over Time')

# Valence over time
plt.subplot(3, 1, 3)
sns.lineplot(x='release_month', y='valence', hue='release_year', data=seasonal_data)
plt.title('Valence Over Time')

plt.tight_layout()
plt.show()
```

```python
# Convert 'track_album_release_date' to datetime format
df['track_album_release_date'] = pd.to_datetime(df['track_album_release_date'])

# Extract month and year from the release date
df['release_month'] = df['track_album_release_date'].dt.month_name()
df['release_year'] = df['track_album_release_date'].dt.year

# Create a pie chart for popularity
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
df.groupby('release_month')['track_popularity'].mean().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Average Popularity by Month')

# Create a pie chart for tempo
plt.subplot(1, 3, 2)
df.groupby('release_month')['tempo'].mean().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Average Tempo by Month')

# Create a pie chart for valence
plt.subplot(1, 3, 3)
df.groupby('release_month')['valence'].mean().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Average Valence by Month')

plt.tight_layout()
plt.show()
```
