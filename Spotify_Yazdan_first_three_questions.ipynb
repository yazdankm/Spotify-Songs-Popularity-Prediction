<h2>Number of different genres Over Time<h2>

```python


# Group by release year and genre, then count the occurrences of each genre in each year
genre_count_per_year = df.groupby(['release_year', 'playlist_genre']).size().unstack(fill_value=0)

# Filter out less frequent genres or group them into an 'Other' category
top_genres = genre_count_per_year.sum().nlargest(6).index
other_genres = [genre for genre in genre_count_per_year.columns if genre not in top_genres]

if len(other_genres) > 0:
    genre_count_per_year['Other'] = genre_count_per_year[other_genres].sum(axis=1)
    genre_count_per_year.drop(other_genres, axis=1, inplace=True)

# Create a grid of subplots (2 rows, 3 columns) for each genre
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))
axes = axes.flatten()

# Loop through each top genre and plot it in a separate subplot
for i, genre in enumerate(top_genres):
    ax = axes[i]
    ax.plot(genre_count_per_year.index, genre_count_per_year[genre], label=genre, color='blue')
    ax.set_xlabel('Release Year')
    ax.set_ylabel('Number of Songs')
    ax.set_title(f'Number of {genre} Over Time')
    ax.legend()

# Hide empty subplots if the number of genres is less than 6
for j in range(len(top_genres), 6):
    axes[j].axis('off')

plt.tight_layout()
plt.show()




```

<h2>Average Tempo for Each Month and each year<h2>

```python


# Calculate average tempo over years
average_tempo_by_year = df.groupby('release_year')['tempo'].mean()

# Years to plot average tempo for each month
years_to_plot = [1980, 1991, 2005, 2019]

# Filter the DataFrame for specified years and calculate average tempo for each month
tempo_by_month = df[df['release_year'].isin(years_to_plot)].groupby(['release_year', 'release_month'])['tempo'].mean().unstack()

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12, 10))

# Plotting average tempo over years
axes[0].plot(average_tempo_by_year.index, average_tempo_by_year.values, marker='o', color='blue')
axes[0].set_xlabel('Release Year')
axes[0].set_ylabel('Average Tempo')
axes[0].set_title('Average Tempo Over Years')

# Plotting average tempo for each month for specified years
for year in years_to_plot:
    axes[1].plot(tempo_by_month.loc[year].index, tempo_by_month.loc[year].values, marker='o', label=year)

# Set x-axis ticks and labels to month names
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
axes[1].set_xticks(range(1, 13))
axes[1].set_xticklabels(months)

axes[1].set_xlabel('Month')
axes[1].set_ylabel('Average Tempo')
axes[1].set_title('Average Tempo for Each Month')
axes[1].legend(title='Years', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

<h2>Top 5 Genres by Average Popularity in different decades<h2>

```python


# Create a new column for decades
df['decade'] = (df['release_year'] // 10) * 10

# Grouping data by decade and genre
grouped = df.groupby(['decade', 'playlist_genre'])

# Calculate the average popularity of each genre in each decade
average_popularity = grouped['track_popularity'].mean()

popularity_by_decade = {}

for (decade, genre), popularity in average_popularity.items():
    if decade not in popularity_by_decade:
        popularity_by_decade[decade] = []
    popularity_by_decade[decade].append((genre, popularity))

# Sort genres in each decade by popularity
for decade in popularity_by_decade:
    popularity_by_decade[decade].sort(key=lambda x: x[1], reverse=True)

for decade in sorted(popularity_by_decade):
    genres = [genre for genre, _ in popularity_by_decade[decade][:5]]
    popularities = [popularity for _, popularity in popularity_by_decade[decade][:5]]
    
    # Create a colormap with shades from darker to lighter based on popularity
    cmap = plt.cm.get_cmap('Blues')
    colors = cmap(np.linspace(0.2, 1, len(genres)))  # Adjust the range for better color differentiation
    colors = colors[::-1]  # Reverse the order of colors
    
    plt.figure(figsize=(10, 6))
    bars = plt.barh(genres, popularities, color=colors)
    plt.ylabel('Genre')
    plt.xlabel('Average Popularity')
    plt.title('Top 5 Genres by Average Popularity in the ' + str(decade) + 's')
    plt.gca().invert_yaxis()
    plt.show()
```

<h2>Monthly and yearly Trends for track_popularity, tempo, valence<h2>

```python
# Filter the DataFrame for the specific years
years_to_plot = [1970, 1985, 2000, 2019]
df_filtered_years = df[df['release_year'].isin(years_to_plot)]

# Group by year and calculate average popularity, tempo, and valence
yearly_data = df_filtered_years.groupby('release_year').mean()

# Group by year and month and calculate average popularity, tempo, and valence
monthly_data = df_filtered_years.groupby(['release_year', 'release_month']).mean()

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot trends over the years for popularity, tempo, and valence
for col in ['track_popularity', 'tempo', 'valence']:
    axes[0].scatter(yearly_data.index, yearly_data[col], marker='o', label=col.capitalize())
    axes[0].plot(yearly_data.index, yearly_data[col], linestyle='-', alpha=0.5)
    axes[0].set_xlabel('Year')
    axes[0].set_ylabel('Average')
    axes[0].set_title('Yearly Trends')
    axes[0].legend()

# Plot trends over the months for popularity, tempo, and valence for the specified years
years_to_plot = [1970, 2019]
for year in years_to_plot:
    monthly_data_year = monthly_data.loc[year]
    for col in ['track_popularity', 'tempo', 'valence']:
        axes[1].plot(monthly_data_year.index.get_level_values('release_month'), monthly_data_year[col], marker='o', label=f'{year} - {col.capitalize()}')
        axes[1].set_xlabel('Month')
        axes[1].set_ylabel('Average')
        axes[1].set_title('Monthly Trends')
        axes[1].legend()
        axes[1].set_xticks(range(1, 13))  # Set x-ticks from 1 to 12 (January to December)
        axes[1].set_xticklabels(['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'], fontsize=8, rotation=90)  # Adjust font size and rotate labels vertically

plt.tight_layout()
plt.show()
```
