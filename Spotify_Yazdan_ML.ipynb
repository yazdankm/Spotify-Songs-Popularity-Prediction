```python
# Create a function to define popularity based on track_popularity values
def define_popularity(row):
    if row['track_popularity'] < 10:
        return 'low'
    elif 10 <= row['track_popularity'] < 50:
        return 'medium'
    elif 50 <= row['track_popularity'] < 90:  # Changed else to elif
        return 'high'
    else:
        return 'highest'

# Apply the function to create the 'popularity' column
df['popularity'] = df.apply(define_popularity, axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
from scipy.optimize import curve_fit
from sklearn.preprocessing import FunctionTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder



label_encoder = LabelEncoder()
df['artist'] = label_encoder.fit_transform(df['track_artist'].astype(str))

label_encoder = LabelEncoder()
df['subgenre'] = label_encoder.fit_transform(df['playlist_subgenre'].astype(str))



# Splitting the data into features (X) and target variable (y)
X = df[['artist','subgenre','release_year', 'release_month', 'release_day','danceability','energy', 'loudness', 'speechiness', 'acousticness','instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms']]
y = df['popularity']



# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)



# Transform features by scaling specific features to a given range
ctr = ColumnTransformer([('minmax', MinMaxScaler(), ['artist','subgenre','release_year', 'release_month', 'release_day', 'duration_ms']),
                        ('categorical', OneHotEncoder(), ['key'])],
                       remainder='passthrough')



# Initialize models
models = {
    'Decision Tree': DecisionTreeClassifier(),
    'KNN': KNeighborsClassifier(),
    'Logistic Regression': LogisticRegression(),
    'Random Forest': RandomForestClassifier()
}

# Train and evaluate models
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    print(f"Model: {name}")
    print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
    print(classification_report(y_test, y_pred))
    print("-" * 50)
```
