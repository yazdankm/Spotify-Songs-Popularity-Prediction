```python
#CHANGES IN DATE COLUMN 
```

```python
my_df = df.copy()
```

```python
my_df['release_year'] = pd.DatetimeIndex(df['track_album_release_date']).year
my_df['release_month'] = pd.DatetimeIndex(df['track_album_release_date']).month
my_df['release_day'] = pd.DatetimeIndex(df['track_album_release_date']).day
```

```python
column_order = ['track_id', 'track_name', 'track_artist', 'track_album_id', 
                 'track_album_name', 'track_album_release_date', 
                 'release_year', 'release_month', 'release_day',
                 'playlist_name', 'playlist_id', 'playlist_genre', 'playlist_subgenre',
                 'danceability', 'energy', 'key', 'loudness', 'mode',
                 'speechiness', 'acousticness', 'instrumentalness', 'liveness',
                 'valence', 'tempo', 'duration_ms']

my_df = my_df[column_order]
```
